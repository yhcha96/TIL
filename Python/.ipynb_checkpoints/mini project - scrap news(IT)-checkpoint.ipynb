```python
################################## 기사 모으기 ##################################
import requests
from bs4 import BeautifulSoup
import urllib.request as req
import re

all_news = []
for i in range(40):
    page = str(i+1)
    url = "https://www.asiae.co.kr/list/IT/"+ page
    response = requests.get(url)
    text = response.text
    soup = BeautifulSoup(text,'html.parser')
    
    links = []
    for link in soup.select("h3.l_fsttit a"):
        links.append(link.attrs['href']) # 각 뉴스 url 추출
    
    
    for link in links:
        news = ''
        url = "http:" + link # 뉴스 url 
        response = requests.get(url)
        text = response.text
        soup = BeautifulSoup(text,'html.parser')
        
        for x in soup.select("div#txt_area p"):
            if re.search("빅데이터|인공지능",x.text): # 빅데이터 또는 인공지능이라는 단어가 있는지 체크
                for x in soup.select("div#txt_area p"):
                    news += x.text # 뉴스에 있는 본문이 추출
                all_news.append(news) # 뉴스 저장
                break
                
################################## 리스트를 텍스트 파일로 저장 ###########################
f = open('newsBigData.txt','w')
for i in all_news:
    f.write(i)
f.close()

c = open('newsBigData.txt').read() # 텍스트 파일 읽기

########################## 형태소 분석 #######################################
from konlpy.tag import Mecab
from nltk import Text

mecab = Mecab('C:\mecab\mecab-ko-dic')
news_nouns = Text(mecab.nouns(c)) # mecab으로 명사만 추출

# fd = news_nouns.vocabcab()
# fd.most_common(30) # 가장 많이 나온 상위 30개 단어

################################ wordcloud ##########################################
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 자신의 컴퓨터 환경에 맞는 한글 폰트 경로를 설정
font_path = "C:\Windows\Fonts\malgun.ttf"

wc = WordCloud(width = 1000, height = 600, background_color="white", font_path=font_path)
plt.imshow(wc.generate_from_frequencies(news_nouns.vocab()))
plt.axis("off")
plt.show()
```
