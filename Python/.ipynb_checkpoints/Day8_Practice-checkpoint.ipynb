```python
# ID : [\w.-]+
# address : ([a-zA-Z]+[.][a-zA-Z]+)|([a-zA-Z]+[.][a-zA-Z]+[.][a-zA-Z]+)
import re

p = re.compile("[\w.+-]+[@]([a-zA-Z-]+[.][a-zA-Z]+)|[\w.+-]+[@]([a-zA-Z-]+[.][a-zA-Z]+[.][a-zA-Z]+)")
email = input()
if p.match(email):
    print("This is correct email form")
else:
    print("This is wrong email form")
```

```python
2. 토지 파일 전처리(정규식)
-저자명 추출
-편 제목 추출
ex) 제2편 제목 - 추적과 음모
-각 편에 대한 장 구성 정보 출력
ex) 제 1편 - ??장으로 구성
-인용부호(겹따옴표)로 묶여있는 대사만 추출->챗봇
 (토지/어린왕자 대사 추출)
-치환(양반, 사람 => 사람)
```

```python
# 저자명 <author>뒤에 있음
import re
f = open('c:/my_Share/BEXX0003.txt','r',encoding = "UTF-16")

while True:
    line = f.readline()
    pat = re.search("<author>([가-힣]+)</author>",line)
    if pat:
        print(pat.group(1))
        break
    if not line:
        break
f.close()
```

```python
# 편제목 <head>
import re
f = open('c:/my_Share/BEXX0003.txt','r',encoding = "UTF-16")

while True:
    line = f.readline()
    pat = re.search("<head>제(\d)편 ([가-힣\s]+)</head>",line)
    if pat:
        print("제 {}편은 {}입니다".format(pat.group(1),pat.group(2)))
    if not line:
        break
f.close()
```

```python
# 장 <head>
import re
f = open('c:/my_Share/BEXX0003.txt','r',encoding = "UTF-16")

c=[]
t=0
while True:
    line = f.readline()
    pat = re.search("<head>(\d)장",line)
    if pat:
        c += [int(pat.group(1))]
    if not line:
        break

for i in range(1, len(c)):
    if c[i-1] > c[i]:
        t += 1
        print("{}편의 장 수는 {}입니다".format(t,c[i-1]))
    if i == len(c)-1:
        t += 1
        print("{}편의 장 수는 {}입니다".format(t,c[i]))

f.close()
```

```python
#인용부호
import re
f = open('c:/my_Share/BEXX0003.txt','r',encoding = "UTF-16")

i = 0
while True:
    line = f.readline()
    pat = re.search("<p>\"([가-힣\s\,\.\?\!\-\"]+)\"</p>",line)
    if pat:
        print(pat.group(1))
        i += 1
    if not line:
        break
f.close()
print(i)
```

```python
#치환
import re
f = open('c:/my_Share/BEXX0003.txt','r',encoding = "UTF-16")

data = []
while True:
    line = f.readline()
    pat = re.search("양반|사람",line)
    if pat:
        a = re.sub("양반|사람","사람",line)
        data += [a]
    else:
         data +=[line]
    if not line:
        break
f.close()

ff = open("c:/my_Share/test_togi.txt",'w')
for i in data:
    ff.write(i)

ff.close()
```
