```python
import pandas as pd
import re

class PhoneBook:
    def __init__(self):
        data = pd.DataFrame(columns=['Name','Phone Number','E-mail','Address'])
        data.to_csv("Phone Book")
        PhoneBook.main(self)
        
    def main(self):
        print('1. 연락처 입력','2. 연락처 출력','3. 연락처 삭제','4. 종료',end='\n')
        choice = input('메뉴선택: ')
        
        if choice == '1':
            PhoneBook.add(self)
        elif choice == '2':
            PhoneBook.show(self)
        elif choice == '3':
            PhoneBook.delete(self)
        elif choice == '4':
            PhoneBook.end(self)
        else:
            print('잘못된 입력을 하였습니다. 다시 선택 해주세요(1~4)')
            PhoneBook.main(self) # 함수를 호출할때마다 함수내에 존재하는 것들이 메모리에 있게되어 while문을 사용하는 걸 추천 
                                # 함수호출을 최대한 적게하는게 좋음
            
            
    def add(self):        
        # csv파일 읽어오기
        data = pd.read_csv("Phone Book")
        
        Name = input("Name: ")
        Phone = input("Phone Number: ")
        Email = input("E-mail: ")
        Address = input("Address: ")
        
        n = re.match('[가-힣]{2,4}',Name)
        p = re.match('\d{3}[-]\d{4}[-]\d{4}|\d{2}[-]\d{3}[-]\d{4}',Phone)
        e = re.match('[0-9a-zA-Z][_0-9a-z.A-Z-]*@[_0-9a-zA-Z-]+(\.[_0-9a-zA-Z-]+){1,2}$',Email)
        a = re.match('[\w가-힣-\s]+',Address)
        
        
        if n and p and e and a:
            # 파일에 저장 DataFrame으로 저장
            new = pd.DataFrame([[Name,Phone,Email,Address]],columns=['Name','Phone Number','E-mail','Address'])
            data = pd.concat([data,new],axis=0,ignore_index=True)
        else:
            print('잘못 입력된 정보가 있습니다. 다시 입력해주세요')
        
        # csv파일로 저장
        data.to_csv("Phone Book")
            
        return PhoneBook.main(self)
    
    
    def delete(self): 
        # csv파일 읽어오기
        data = pd.read_csv("Phone Book")
        
        # 4개 입력 받기
        find = input("삭제할 번호의 4자리를 입력해주세요")
        delete_list=[]
        
        # 4개의 숫자와 맞는 번호 찾기
        if re.match('\d{4}',find):
            for i in range(len(data["Phone Number"])):
                if re.search(find, data["Phone Number"][i]):
                    delete_list.append([i,data["Phone Number"][i]])
        else:
            print('숫자 4개를 입력해주세요 다시 입력')
            PhoneBook.delete(self)
            
                    
        if delete_list:
            # 목록 주고 삭제 선택
            for i in range(len(delete_list)):
                print("{}. {}".format(i,delete_list[i][1]))
            num = int(input('삭제할 전화번호를 선택해주세요'))
            # DataFrame에서 행 삭제
            data = data.drop([delete_list[num][0]])
            
        else:
            print(find,'와 매칭되는 번호가 없습니다.')
        
        # csv파일로 저장
        data.to_csv("Phone Book")
        
        return PhoneBook.main(self)
                    
    
    
    def show(self):
        # csv파일 읽어오기
        data = pd.read_csv("Phone Book")
        if data.empty:
            print("연락처에 저장된 것이 없습니다")
        else:
            for i in list(data.index):
                print("Name: ",data['Name'][i])
                print("Phone Number: ",data["Phone Number"][i])
                print("E-mail: ",data['E-mail'][i])
                print("Address: ",data['Address'][i])
                print('-'*30)
            
        return PhoneBook.main(self)
    
    
    def end(self):
        print("종료합니다")
```
