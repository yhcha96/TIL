```python
import pandas as pd
import numpy as np
```

```python
# zoo.data 읽기
data = pd.read_csv('zoo.data')
```

```python
# 목록 이름이 없기때문에 header를 None로 해주기
data = pd.read_csv('zoo.data',header= None)
```

```python
# zoo.names에서 columns 이름 추출
import re

f = open('zoo.names','r')
lines = f.readlines()

columns=[]
for line in lines:
    m = re.search("\s+\d+[.]\s([\w ]+)",line)
    if m:
        columns.append(m.group(1))
```

```python
# columns에 추출한 columns이름 넣기
data.columns = columns
```

```python
# 새로운 열 만들기
new =[]
for i in data['animal name']:
    if i in ['aardvark', 'antelope', 'bear', 'boar', 'buffalo', 'calf',
                  'cavy', 'cheetah', 'deer', 'dolphin', 'elephant',
                  'fruitbat', 'giraffe', 'girl', 'goat', 'gorilla', 'hamster',
                  'hare', 'leopard', 'lion', 'lynx', 'mink', 'mole', 'mongoose',
                  'opossum', 'oryx', 'platypus', 'polecat', 'pony',
                  'porpoise','puma', 'pussycat', 'raccoon', 'reindeer',
                  'seal', 'sealion', 'squirrel', 'vampire', 'vole', 'wallaby','wolf']:
        new.append(1)
    elif i in ['chicken', 'crow', 'dove', 'duck', 'flamingo', 'gull', 'hawk',
                  'kiwi', 'lark', 'ostrich', 'parakeet', 'penguin', 'pheasant',
                  'rhea', 'skimmer', 'skua', 'sparrow', 'swan', 'vulture', 'wren']:
        new.append(2)
    elif i in ['pitviper', 'seasnake', 'slowworm', 'tortoise', 'tuatara']:
        new.append(3)
    elif i in ['bass', 'carp', 'catfish', 'chub', 'dogfish', 'haddock',
                  'herring', 'pike', 'piranha', 'seahorse', 'sole', 'stingray', 'tuna']:
        new.append(4)
    elif i in ['frog', 'frog', 'newt', 'toad']:
        new.append(5)
    elif i in ['flea', 'gnat', 'honeybee', 'housefly', 'ladybird', 'moth', 'termite', 'wasp']:
        new.append(6)
    elif i in ['clam', 'crab', 'crayfish', 'lobster', 'octopus',
                  'scorpion', 'seawasp', 'slug', 'starfish', 'worm']:
        new.append(7)

# 새로운 열만들어서 new값 넣기        
data[18]=new
# 새로만든 열의 이름 추가하기
data.columns = columns+['class']
```

```python
data
```

```python
# fly이라는 열 추가, fill_value로 그 값을 0으로 채우기 안 채우면 NaN
original_columns = columns+['class']
data = data.reindex(columns = original_columns+["fly"],fill_value=0)
```

```python
# csv로 저장하기
# index=False를 해주어야 index를 보여주는 열이 생성되어 저장되지 않는다
data.to_csv('zoo.csv',index=False)
```
