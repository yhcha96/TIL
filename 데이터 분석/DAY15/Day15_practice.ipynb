```python
df12 = pd.merge(df1,df2,how='outer')
df123 = pd.merge(df12,df3,how='outer')
df123
```

```python
# concat에서 axis=0로 상하로 합치기
df123_0 = pd.concat([df1,df2,df3],axis=0)
df123_0
```

```python
# concat에서 axis=1로 좌우 합치기
df123_1 = pd.concat([df1,df2,df3],axis=1)
df123_1
```

```python
# ignore_index
df123_0_jgn = pd.concat([df1,df2,df3],axis=0,ignore_index=True)
df123_0_jgn
```

```python
df123_1_jgn = pd.concat([df1,df2,df3],axis=1,ignore_index=True)
df123_1_jgn
```

```python
df123_0_key = pd.concat([df1,df2,df3],axis=0,keys=['df_1','df_2','df_3'])
df123_0_key
```

```python
df123_1_key = pd.concat([df1,df2,df3],axis=1,keys=['df_1','df_2','df_3'])
df123_1_key
```

```python
df123_0_key = pd.concat([df1,df2,df3],axis=0,keys=['df_1','df_2','df_3'],names=['dfName','rowData'])
df123_0_key
```

```python
df123_1_key = pd.concat([df1,df2,df3],axis=1,keys=['df_1','df_2'],names=['dfName','columnData'])
df123_1_key
```
