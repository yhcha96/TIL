### join_axes

```python
pd.concat([df1,df4],axis=1,join='inner',join_axes=[df1.index])
# df1과 df4를 좌우로 합친다(axis=1)
# 행인텍스가 공통적으로 존재하는데 데이터에 대해 (join='inner')
# df1에 있는 모든 행인텍스가 출력 df4는 공통적이것만 출력 (join_axes=[df1.index])
```

### verify_integrity
- index 중복 여부 확인
- default False
- True : index 중복시 에러 발생

```python
df7 = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
     'B': ['B0', 'B1', 'B2'],
     'C': ['C0', 'C1', 'C2'],
     'D': ['D0', 'D1', 'D2']},
     index=['r0', 'r1', 'r2'])

df8 = pd.DataFrame({'A': ['A2', 'A3', 'A4'],
     'B': ['B2', 'B3', 'B4'],
     'C': ['C2', 'C3', 'C4'],
     'D': ['D2', 'D3', 'D4']},
     index=['r2', 'r3', 'r4'])
```

```python
pd.concat([df7,df8],verify_integrity=False) # index에 r2가 중복이 되지만 에러발생x

# pd.concat([df7,df8],verify_integrity=True) # index에 r2가 중복임으로 엘러 발생
```

```python
pd.concat([df7,df8],verify_integrity=True, ignore_index=True) # ignore_index로 index를 reset했기 때문에 에러발생x
```

```python
# pd.concat([df7,df8],axis =1, verify_integrity=True) # ABCDABCD임으로 중복되어 에러발생
pd.concat([df7,df8],axis =1, verify_integrity=True,ignore_index=True)
```

### 3. Merge
- common colmn이 있어야 가능
- how : default - inner
- on 
- indicator='indicator_info' : 데이터가 어디서 왔는지 알려줌
- suffixes

```python
dfright = DataFrame({'KEY': ['K2', 'K3', 'K4', 'K5'],
                     'C': ['C2', 'C3', 'C4', 'C5'],
                     'D': ['D2', 'D3', 'D4', 'D5']})

   
dfleft = DataFrame({'KEY': ['K0', 'K1', 'K2', 'K3'],
                    'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3']})
```

```python
# merge는 공통된 것을 기준으로 합쳐진다
pd.merge(dfleft,dfright)
```

### how

```python
# defalut : inner
pd.merge(dfleft,dfright,how="outer")
```

### on

```python
pd.merge(dfleft,dfright,how="left",on='KEY')
```

```python
pd.merge(dfleft,dfright,how="left",on='KEY') # on='KEY' KEY를 기준으로, how='left' index는 left기준
```

```python
pd.merge(dfleft,dfright,how="left")
```

```python
pd.merge(dfleft,dfright,how="right",on='KEY') # dfright의 KEY가 값이 기준
```

```python
pd.merge(dfleft,dfright,how="right")
```

### indicator

```python
#inner교집합 outer합집합
pd.merge(dfleft,dfright,how='outer', on='KEY',indicator='indicator_info')
```

### suffixes

```python
dfleft2 = DataFrame({'KEY': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3'],
                     'C': ['C0', 'C1', 'C2', 'C3']})

dfright_2 = DataFrame({'KEY': ['K0', 'K1', 'K2', 'K3'],
                       'B': ['B0_2', 'B1_2', 'B2_2', 'B3_2'],
                       'C': ['C0_2', 'C1_2', 'C2_2', 'C3_2'],
                       'D': ['D0_2', 'D1_2', 'D2_2', 'D3_3']})
```

```python
pd.merge(dfleft2,dfright_2, how='inner', on = 'KEY', suffixes=('_left','_right')) # 중복된거 어디서 온건지 표시
```

```python
pd.merge(dfleft2,dfright_2, on = 'KEY')
```

### 4. 데이터프레임을 인덱스를 기준으로 병합
- merge
- join

```python
dfleft = DataFrame({
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']},
    index=['K0', 'K1', 'K2', 'K3'])


dfright = DataFrame({
    'C': ['C2', 'C3', 'C4', 'C5'],
    'D': ['D2', 'D3', 'D4', 'D5']},
    index= ['K2', 'K3', 'K4', 'K5'])
```

```python
dfleft = DataFrame({
    'A': ['1', 'A1', 'A2', 'A3'],
    'B': ['1', 'B1', 'B2', 'B3']},
    index=['K0', 'K1', 'K2', 'K3'])


dfright = DataFrame({
    'C': ['1', 'C3', 'C4', 'C5'],
    'D': ['1', 'D3', 'D4', 'D5']},
    index= ['K2', 'K3', 'K4', 'K5'])
```

### merge

```python
# pd.merge(dfleft,dfright) # 공통된 열이 없기때문에 에러
pd.merge(dfleft,dfright,left_index=True,right_index=True)
```

```python
pd.merge(dfleft,dfright,left_index=True,right_index=True, how='left')
```

```python
pd.merge(dfleft,dfright,left_index=True,right_index=True,how='right')
```

### join
- how : dealut - left

```python
dfleft.join(dfright) # pd.merge(dfleft,dfright,left_index=True,right_index=True,how='left')
```

```python
dfleft.join(dfright,how='right') # pd.merge(dfleft,dfright,left_index=True,right_index=True,how='right')
```

### 기존 index로 합치기

```python
dfleft = DataFrame({
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']},
    index=['K0', 'K1', 'K2', 'K3'])


dfright = DataFrame({
    'C': ['C2', 'C3', 'C4', 'C5'],
    'D': ['D2', 'D3', 'D4', 'D5']},
    index= ['K2', 'K3', 'K4', 'K5'])
```

```python
# index가 다음처럼 나올려면???
# K0
# K1
# K2
# K3

#merge
# left_index=True,right_index=True 필수 how='left' 안해주면 inner로 됨
print(pd.merge(dfleft,dfright,how='left',left_index=True,right_index=True))

#join
print(dfleft.join(dfright))
```

```python
# index가 K2 K3 K4 K5 이려면??
print(pd.merge(dfleft,dfright,how='right',left_index=True,right_index=True))
print(dfleft.join(dfright,how='right'))
```

```python
# dfleft, dfrignt outer join

print(pd.merge(dfleft,dfright,how='outer',left_index=True,right_index=True))
print(dfleft.join(dfright,how='outer'))
```

### index와 key조합으로 합치기

```python
dfleft = DataFrame({'KEY':['K0', 'K1', 'K2', 'K3'],
                    'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3']})


dfright = DataFrame({'C': ['C2', 'C3', 'C4', 'C5'],
                     'D': ['D2', 'D3', 'D4', 'D5']},
                    index= ['K2', 'K3', 'K4', 'K5'])
```

```python
# pd.merge(dfleft,dfright) # 공통된 열(column) 없으니깐 에러

pd.merge(dfleft,dfright,left_on="KEY",right_index=True, how='left') 
# dfleft의 KEY와 dfright의 index가 같으니깐 left_on='KEY', right_index=True
# how='left' : dfleft의 'KEY'기준
```

```python
dfleft.join(dfright,on='KEY',how='left') 
```
