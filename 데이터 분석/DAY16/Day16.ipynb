```python
df_left= pd.DataFrame({
    'Key':['K0','K1','K2','K3'],
    'A':['A0','A1','A2','A3'],
    'B':[0.5,1.0,0.7,1.5]
})

df_right= pd.DataFrame({
    'Key':['K2','K3','K4','K5'],
    'C':['C0','C1','C2','C3'],
    'D':['D2','D3','D4','D5']
})
```

```python
df_all = pd.merge(df_left,df_right,how='outer',on='Key')
```
